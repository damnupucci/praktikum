Похоже, для разного пола данные похожи

```python
# заполним пропуски среднем стажем, плюс стажем в этом столбце, разделенном на 100
data.loc[data['days_employed'] > 0, 'days_employed'] = data.loc[data['days_employed'] > 0,
                                                                'days_employed'] / 100 + data[data['days_employed'] < 0]['days_employed'].abs().mean()
data.loc[data['days_employed'] < 0,'days_employed']  = data.loc[data['days_employed'] < 0,'days_employed'].abs()
```

### Где и зачем использовать конструкцию try-except?


```python
try:
    data_right = pd.read_csv('data.csv', delimiter='\t')
    data_right.info()
except:
    print("File not found")
```

```python
try:
    data_wrong = pd.read_csv('data_nonexistent_path.csv', delimiter='\t')
    data_wrong.info()
except:
    print("File not found")
```

```python
my_series = pd.Series([1, 2, 3, 4])
try:
    print(my_series.mean())
except:
    print("Error applying mean")
```

```python
lemm_home_dict = {
  'покупка жилья': 'жилье',
  'операции с жильем': 'жилье',
  'покупка жилья для семьи': 'жилье',
  'жилье': 'жилье',
  'покупка своего жилья': 'жилье',
  'покупка жилья для сдачи': 'жилье',
  'ремонт жилью':'жилье'
}

def home_category(some_string):
    try:
        return lemm_home_dict[some_string]
    except:
        return "no_category"
```

```python
home_category('ремонт дома')
```
